```python
def remove_features_with_missing_values(df, threshold=0.2):
    missing_percentage = df.isnull().mean()
    columns_to_drop = missing_percentage[missing_percentage > threshold].index
    df = df.drop(columns=columns_to_drop)

    return df

df = remove_features_with_missing_values(df, threshold=0.2)
df.info()
```

```python
from extract_label import parse_sas_html

sas_variable, value_labels = parse_sas_html("USCODE23_LLCP_091024.HTML")
```

```python
cols = df.columns
for col in cols:
    print(f"{col}")
```

```python

relevant_features = [
    # Continuous variables
    '_AGE80', '_BMI5', 'WTKG3', 'HTM4',

    # Diabete
    'DIABETE4', 'DIABTYPE', '_DRDXAR2',  # Diabetes, kidney, arthritis

    # Chronic health conditions
    'CVDINFR4',  # TARGET VARIABLE
    'CVDCRHD4', 'CVDSTRK3', 'ASTHMA3', 'ASTHNOW', 'CHCSCNC1', 'CHCOCNC1', 'CHCCOPD3', 'ADDEPEV3', 'CHCKDNY2', 'HAVARTH4',

    # Heart attack symptoms
    'HASYMP1', 'HASYMP2', 'HASYMP3', 'HASYMP4', 'HASYMP5', 'HASYMP6',
    
    # Demographics
    '_SEX', '_AGEG5YR', '_RACE',  # Sex, age, and race/ethnicity are important risk factors.

    # Health Status
    '_RFHLTH', '_PHYS14D', '_CHOLCH3', '_RFHYPE6', '_MICHD',   # Health conditions like general health, physical health, cholesterol, blood pressure, heart disease, and current e-cigarette use are key predictors.

    # Lifestyle
    '_SMOKER3', '_CURECI2', '_RFSMOK3', '_PA150R4', '_PA300R4', '_PASTRNG', '_RFBMI5',  # Smoking status, physical activity, and BMI are strongly associated with heart disease.

    # Risk Factors
    '_BMI5CAT', '_RFCHOL3', '_RFBING6', '_RFDRHV8', '_TOTINDA',  # Body mass index (BMI) category, cholesterol levels, alcohol consumption, and physical activity are key risk factors.

    # Health Access & Monitoring
    '_PAINDX3', '_PASTAE3', '_PACAT3'  # physical activity levels, and health checkups play a role in heart disease risk.
]
```

```python
filtered_df = df[[col for col in relevant_features if col in df.columns]]

print(filtered_df.head())
print(f"Number of features after filtering: {len(filtered_df.columns)}")
```
