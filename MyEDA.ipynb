```python
import scipy.stats as stats

# Function to compute Cramer's V
def cramers_v(contingency_table):
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()  # Total number of observations
    phi2 = chi2 / n
    r, k = contingency_table.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

# List of categorical variables (adjust this list to your dataset)
categorical_columns = categorical_cols

# Impute missing data with mode for categorical columns
for col in categorical_columns:
    mode_value = df[col].mode()[0]  # Get the mode value for each column
    df[col].fillna(mode_value, inplace=True)  # Fill missing values with mode

# Initialize an empty correlation matrix for Cramer's V
cramers_v_matrix = pd.DataFrame(np.nan, index=categorical_columns, columns=categorical_columns)

# Loop through all pairs of categorical variables (only half of the matrix needs to be filled)
for i, col1 in enumerate(categorical_columns):
    for j, col2 in enumerate(categorical_columns):
        if i < j:  # Process each pair once
            # Create contingency table
            contingency_table = pd.crosstab(df[col1], df[col2])
            
            # Perform Chi-Square test
            chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
            
            # Calculate Cramer's V
            cramer_v = cramers_v(contingency_table)
            
            # Store Cramer's V in the matrix
            cramers_v_matrix.loc[col1, col2] = cramer_v
            cramers_v_matrix.loc[col2, col1] = cramer_v  # Fill symmetric part of the matrix
```

```python
plt.figure(figsize=(20, 18))
sns.heatmap(cramers_v_matrix.astype(float), annot=False, cmap='coolwarm', vmin=0, vmax=1)
# Adjust labels and ticks for better readability
plt.title("Cramér's V Heatmap for Categorical Variables", fontsize=14)

# Rotate the x and y axis labels and adjust the font size
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and set font size
plt.yticks(rotation=0, ha='right', fontsize=10)   # Keep y labels horizontal and adjust font size

plt.tight_layout()  # Adjust layout to avoid overlapping of elements

# Show the plot
plt.show()
```

```python
cramers_v_for_target = cramers_v_matrix.loc['CVDINFR4'].dropna()

# Filter and display predictors with Cramer's V greater than 0.5
high_cramers_v_for_target = cramers_v_for_target[cramers_v_for_target > 0.6]

print("Categorical variables with Cramer's V > 0.5 for CVDINFR4:")
for var, cramer_v in high_cramers_v_for_target.items():
    print(f"{var}: {cramer_v:.2f}")
```

```python
high_cramers_v_pairs = cramers_v_matrix.where(cramers_v_matrix > 0.6)

# Drop the diagonal and self-comparisons (which will be NaN) and filter only the pairs
high_cramers_v_pairs = high_cramers_v_pairs.stack().reset_index()

# Rename columns for better readability
high_cramers_v_pairs.columns = ['Variable 1', 'Variable 2', "Cramer's V"]

# Sort the variable names in each pair and drop duplicate pairs
high_cramers_v_pairs['pair'] = high_cramers_v_pairs.apply(lambda x: tuple(sorted([x['Variable 1'], x['Variable 2']])), axis=1)

# Remove duplicates based on the sorted pair
high_cramers_v_pairs = high_cramers_v_pairs.drop_duplicates(subset='pair')
high_cramers_v_pairs = high_cramers_v_pairs.sort_values(by="Cramer's V", ascending=False)
# Display pairs with Cramer's V > 0.5
print("Pairs of categorical variables with Cramér's V > 0.5:")

# Iterate through the rows of the dataframe and print pairs
for idx, row in high_cramers_v_pairs.iterrows():
    print("({}, {}): Cramér's V = {:.2f}".format(row['Variable 1'], row['Variable 2'], row["Cramer's V"]))

# Print the count of pairs
count = high_cramers_v_pairs.shape[0]
print("\nTotal number of variable pairs with Cramér's V > 0.6: ", count)
```
